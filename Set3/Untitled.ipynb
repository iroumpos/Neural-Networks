```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f(x, y) and f(y, z)
def f_xy(x, y, k):
    return np.exp(-k * (x - y)**2)

def f_yz(y, z, k):
    return np.exp(-k * (y - z)**2)

# Define the range for x, y, and z
x_values = np.linspace(-1, 1, 100)
y_values = np.linspace(-1, 1, 100)
z_values = np.linspace(-1, 1, 100)

# Create a meshgrid for x, y, and z
X, Y = np.meshgrid(x_values, y_values)
Y, Z = np.meshgrid(y_values, z_values)

# Choose a value for k
k = 1

# Calculate the functions
fxy_values = f_xy(X, Y, k)
fyz_values = f_yz(Y, Z, k)

# Create 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface for f(x, y)
ax.plot_wireframe(X, Y, fxy_values, cmap='Blues', alpha=0.8)

# Plot surface for f(y, z)
ax.plot_wireframe(X, Y, fyz_values, cmap='Reds', alpha=0.8)

# Set labels and title
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_title('3D Plot of Functions $f(x, y)$ and $f(y, z)$')
plt.savefig("prob10.png")
# Show the plot
plt.show()
```
